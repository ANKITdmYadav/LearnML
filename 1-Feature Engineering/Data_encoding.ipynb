```python
# Nominal/One Hot Encoding
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
df=pd.DataFrame({
    'color':['red','blue','green','yellow','red','green']
})
encoder=OneHotEncoder()
encoded=encoder.fit_transform(df[['color']]).toarray()
encoded
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
# encoder_df
encoder.transform([['blue']]).toarray()

pd.concat([df,encoder_df],axis=1)
```

```python
import seaborn as sns
df2=sns.load_dataset('titanic')
df2.head()
```

```python
#Assignment
# if sparse_output = true gives sparse matrix output to save memory and we don't want that we want array
encoder2 = OneHotEncoder(sparse_output=False)
sex_encoded = encoder2.fit_transform(df2[['sex']])
encoded_df2=pd.DataFrame(sex_encoded,columns=encoder2.get_feature_names_out())
encoded_df2
# df['encoded_sex']=sex_encoded

# add this new data to df2
new_df2=pd.concat([df2,encoded_df2],axis=1)
new_df2
```

```python
encoder3=OneHotEncoder(sparse_output=False,drop=None)
smoke_encoded=encoder3.fit_transform(df2[['class']])
encoded_df3=pd.DataFrame(smoke_encoded,columns=encoder3.get_feature_names_out())

new_df3=pd.concat([df2,encoded_df3],axis=1)
new_df3
```

LABEL ENCODING


```python
from sklearn.preprocessing import LabelEncoder

df=pd.DataFrame({
    'color':['red','blue','green','yellow','red','green']
})

lbl_enco=LabelEncoder()
lbl_enco.fit_transform(df[['color']])
lbl_enco.transform([['blue']])
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['small','medium','big']])
df=pd.DataFrame({
    'size':['small','medium','big','small']
})
oe.fit_transform(df[['size']])

```
