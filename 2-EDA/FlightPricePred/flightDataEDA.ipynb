```python
# time to minutes
# Step 0: Fill NaNs with empty strings to avoid split errors
df['time_taken'] = df['time_taken'].fillna('')

# Step 1: Split time string
# time_split = df['time_taken'].str.split()

# Step 2: Extract hour and minute parts safely
df['hour_taken'] = df['time_taken'].str.split().str[0].str.split('h').str[0].fillna('0')
df['min_taken'] = df['time_taken'].str.split().str[1].str.split('m').str[0].fillna('0') if df['time_taken'].str.split().str.len().max() > 1 else '0'

# Step 3: Convert to integers, handling possible non-numeric entries
df['hour_taken'] = pd.to_numeric(df['hour_taken'], errors='coerce').fillna(0).astype(int)
df['min_taken'] = pd.to_numeric(df['min_taken'], errors='coerce').fillna(0).astype(int)

# Step 4: Calculate total minutes
df['total_min'] = df['hour_taken'] * 60 + df['min_taken']

df.drop('time_taken',axis=1,inplace=True)


```

```python
# Step 1: Clean up the column (replace actual column name accordingly)
df['stops_info'] = df['stop'].fillna('').str.strip()

# Step 2: Extract number of stops
df['no_of_stops'] = df['stops_info'].apply(
    lambda x: 0 if 'non-stop' in x.lower() else (2 if '2' in x else 1)
)

# Step 3: Extract via city (split on 'Via' and take the part after it, if present)
df['via_city'] = df['stops_info'].apply(
    lambda x: x.split('Via')[-1].strip() if 'Via' in x else None
)
```
